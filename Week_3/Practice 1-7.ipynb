### Practice 1

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sstext.jpg', 0)
rows, cols = img.shape

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

M = np.float32([[1, 0, 100], [0, 1, 50]])
trans1 = cv.warpAffine(img, M, (cols, rows))

# Plot the translated image
plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(trans1)

cv.waitKey(0)
cv.destroyAllWindows()
```

### Practice 2

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sstext.jpg')

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

rows, cols = img.shape[:2]

M = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
reflected_img = cv.warpPerspective(img, M, (int(cols), int(rows)))

# Plot the reflected image
plt.subplot(1, 2, 2)
plt.title("Reflected Image")
plt.imshow(reflected_img)

cv.imwrite("reflection_out.jpg", reflected_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Practice 3

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sstext.jpg')
rows, cols = img.shape[:2]

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Apply rotation transformation
img_rotation = cv.warpAffine(img, 
                cv.getRotationMatrix2D((cols/2, rows/2), 30, 0.6), 
                (cols, rows))

# Plot the rotated image
plt.subplot(1, 2, 2)
plt.title("Rotated Image")
plt.imshow(img_rotation)

cv.imwrite("rotation_out.jpg", img_rotation)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sstext.jpg', 0)
rows, cols = img.shape

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Shear transformation matrix for X-axis
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv.warpPerspective(img, M, (int(cols * 1.5), int(rows * 1.5)))

# Plot the sheared image
plt.subplot(1, 2, 2)
plt.title("Sheared Image")
plt.imshow(sheared_img)

cv.imshow("img", sheared_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

img = cv.imread('img/sstext.jpg', 0)
rows, cols = img.shape

# Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)

# Shear transformation matrix for Y-axis
M = np.float32([[1, 0, 0], [0.5, 1, 0], [0, 0, 1]])
sheared_img = cv.warpPerspective(img, M, (int(cols * 1.5), int(rows * 1.5)))

# Plot the sheared image
plt.subplot(1, 2, 2)
plt.title("Sheared Image Y-Axis")
plt.imshow(sheared_img)

cv.imshow("img", sheared_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Practice 7

```python
import cv2

# Read the input image
image = cv2.imread('img/sstext.jpg')

# Display the original image
cv2.imshow('Original Image', image)
cv2.waitKey(0)

# Apply Gaussian Blur
gaussian = cv2.GaussianBlur(image, (7, 7), 0)
cv2.imshow('Gaussian Blurring', gaussian)
cv2.waitKey(0)

# Apply Median Blur
median = cv2.medianBlur(image, 5)
cv2.imshow('Median Blurring', median)
cv2.waitKey(0)

# Apply Bilateral Filter
bilateral = cv2.bilateralFilter(image, 9, 75, 75)
cv2.imshow('Bilateral Blurring', bilateral)
cv2.waitKey(0)

# Close all windows
cv2.destroyAllWindows()
```
