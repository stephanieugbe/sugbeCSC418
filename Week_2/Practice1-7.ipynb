### Practice 1 

```python
import cv2

image = cv2.imread('img/rgb.png')

#the split method seperate color spaces
B, G, R = cv2.split(image)

# Corresponding channels are seperated
cv2.imshow("original", image)
cv2.waitKey(0)

cv2.imshow("Blue", B)
cv2.waitKey(0)

cv2.imshow("Green", G)
cv2.waitKey(0)

cv2.imshow("Red", R)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

### Practice 2

```python
import cv2

#path to image 1 & 2
image1 = cv2.imread('img/sstext.jpg')
image2 = cv2.imread('img/sstfoyer.jpeg')


image1 = cv2.resize(image1, (500,400))
image2 = cv2.resize(image2, (500,400))

#addweighted with applied parameters
addImage = cv2.addWeighted(image1, 0.5, image2, 0.6, 0)

#window showing output image
cv2.imshow('Weighted Image', addImage)

#De-allocate any associated memory usage
if cv2.waitKey(0) & 0xff == 27:
    cv2.destroyAllWindows()
```

### Practice 3

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/sstfoyer.jpeg')

#plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

#adjust brightness and contrast
brightness = 5
contrast = 1.5
image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness)

#save the image
cv2.imwrite('img/contraststeffsimage.jpeg', image2)
#plot the contrast image
plt.subplot(1, 2, 2)
plt.title("Brightness & Contrast")
plt.imshow(image2)
plt.show()
```

### Practice 4

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/sstfoyer.jpeg')

#plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

#create the sharpening kernel
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])

#sharpen the image
sharpened_image = cv2.filter2D(image, -1, kernel)

#save the image
cv2.imwrite('img/sharpenedsteffsimage.jpg', sharpened_image)

#plot the sharpened image
plt.subplot(1, 2, 2)
plt.title("Sharpening")
plt.imshow(sharpened_image)
plt.show()
```

### Practice 5

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/sstfoyer.jpeg')

#plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

#remove noise using median filter
filtered_image = cv2.medianBlur(image, 15)

#save the image
cv2.imwrite('img/medianblursteffsimage.jpeg', filtered_image)

#plot the contrast image
plt.subplot(1, 2, 2)
plt.title("Median Blur")
plt.imshow(filtered_image)
plt.show()
```

### Practice 6

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/sstfoyer.jpeg')

#plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

#scale the image by a factor of 2 along both axes
scaled_image = cv2.resize(image, None, fx=2, fy=2)

#save the image
cv2.imwrite('img/scaledsteffsimage.jpeg', scaled_image)

#plot the contrast image
plt.subplot(1, 2, 2)
plt.title("scaled")
plt.imshow(scaled_image)
plt.show()
```

### Practice 7

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

image = cv2.imread('img/sstfoyer.jpeg')

#plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(image)

# inverse by subtracting from 255
inverse_image = 255 - image

#save the image
cv2.imwrite('img/inversedsteffsimage.jpeg', inverse_image)

#plot the contrast image
plt.subplot(1, 2, 2)
plt.title("Inverse Color")
plt.imshow(inverse_image)
plt.show()
```
